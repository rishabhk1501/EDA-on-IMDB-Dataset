```python
def convert_string_to_list(str):
  return eval(str)
```

```python
# We will Apply the above function to three columns: genres, cast and crew. Since these columns have null values, we will get rid of null values
imdb_df.loc[~imdb_df['genres'].isna(), 'genres'] = imdb_df.loc[~imdb_df['genres'].isna(), 'genres'].apply(convert_string_to_list)
imdb_df.loc[~imdb_df['cast'].isna(), 'cast'] = imdb_df.loc[~imdb_df['cast'].isna(), 'cast'].apply(convert_string_to_list)
imdb_df.loc[~imdb_df['crew'].isna(), 'crew'] = imdb_df.loc[~imdb_df['crew'].isna(), 'crew'].apply(convert_string_to_list)
```

# 4. Make a table of all the producers and directors of each movie. Find the top 3 producers who have produced movies with the highest average RoI?

```python
# Create a dataframe in which we have no null observations for the column crew
imdb_df_no_na_crew = imdb_df_new[~imdb_df_new['crew'].isna()]
```

```python
imdb_df_no_na_crew.iloc[0]['crew']
```

```python
# Creating a function which returns the list of producer for each movie
def get_producers_and_directors(crew_list):
  try:

    producer_list = []
    director_list = []
    for elem in crew_list:
      if elem['job']=='Producer':
        producer_list.append(elem['name'])
      if elem['job']=='Director':
        director_list.append(elem['name'])
    return producer_list,director_list
  except:
    return [],[]
```

```python
#Check
get_producers_and_directors(imdb_df_no_na_crew.loc[600,'crew'])
```

```python
# Creating two new columns Producers and Directors
imdb_df_new[['Producers', 'Directors']] = imdb_df_new.apply(lambda row: get_producers_and_directors(row['crew']), axis=1, result_type='expand')
```

```python
imdb_df_new.head()
```

```python
movie_prod_dir_df = imdb_df_new[['original_title', 'Producers', 'Directors']]
```

```python
movie_prod_dir_df.head()
```

```python
movie_prod_roi_df = imdb_df_new[['original_title', 'Producers', 'roi']]
```

```python
movie_prod_roi_df.head()
```

```python
movie_prod_roi_df.head()
```

```python
full_producer_list = []
for elem in movie_prod_roi_df['Producers']:
  full_producer_list = full_producer_list + elem
unique_producer_set = set(full_producer_list)
unique_producer_list = list(unique_producer_set)
producer_df = pd.DataFrame(unique_producer_list, columns = ['Producer Name'])
```

```python
producer_df.head()
```

```python
producer_df_copy = producer_df.copy()
producer_df_copy.head()
```

```python
# Function to create
def get_movie_list_for_producer(producer_name):
  try:

    producer_movie_list = []
    for index,row in movie_prod_roi_df.iterrows():
      for elem in row['Producers']:
        if producer_name in elem:
          producer_movie_list.append(row['original_title'])
    return producer_movie_list
  except:
    return []
```

```python
producer_df_copy['movies'] = producer_df_copy['Producer Name'].apply(get_movie_list_for_producer)
```

```python
producer_df_copy.head()
```

```python
producer_df_copy.set_index('Producer Name', inplace=True)
```

```python
producer_df_copy.head()
```

```python
#Check
producer_df_copy.loc['Vin Diesel']['movies']
```

```python
movie_prod_roi_df.set_index('original_title', inplace=True)
```

```python
def get_avg_roi(movie_list):
  sum_roi = 0
  for elem in movie_list:
    sum_roi = sum_roi + (movie_prod_roi_df.loc[elem]['roi'])
    avg_roi = (sum_roi)/len(movie_list)
  return round(avg_roi,2)
```

```python
#Check
get_avg_roi(producer_df_copy.loc['Vin Diesel']['movies'])
```

```python
producer_df_copy['AVG ROI'] = producer_df_copy['movies'].apply(get_avg_roi)
```

```python
producer_df_copy
```

```python
producer_df_copy['AVG ROI'] = pd.to_numeric(producer_df_copy['AVG ROI'], errors='coerce')
```
