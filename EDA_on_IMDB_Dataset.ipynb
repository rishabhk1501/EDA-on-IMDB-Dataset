#Converting the string values to proper lists

```python
# Lets keep only the columns which are required in analysis
```

```python
imdb_df.columns
```

```python
columns_to_keep = ['budget', 'genres', 'imdb_id', 'original_language', 'original_title', 'cast', 'crew', 'revenue']
```

```python
imdb_df = imdb_df[columns_to_keep]
```

```python
imdb_df.head()
```

```python
imdb_df.info()
```

```python
def convert_string_to_list(str):
  return eval(str)
```

```python
# We will Apply the above function to three columns: genres, cast and crew. Since these columns have null values, we will get rid of null values
imdb_df.loc[~imdb_df['genres'].isna(), 'genres'] = imdb_df.loc[~imdb_df['genres'].isna(), 'genres'].apply(convert_string_to_list)
imdb_df.loc[~imdb_df['cast'].isna(), 'cast'] = imdb_df.loc[~imdb_df['cast'].isna(), 'cast'].apply(convert_string_to_list)
imdb_df.loc[~imdb_df['crew'].isna(), 'crew'] = imdb_df.loc[~imdb_df['crew'].isna(), 'crew'].apply(convert_string_to_list)
```

```python
# Creating a copy of the imdb_df
imdb_df_new = imdb_df.copy()
```

#1. Which movie made the highest profit? Who were its producer and director? Identify the actors in that film.

```python
imdb_df_new.describe()
```

```python
# Handling invalid observations
# As we can see budget and revenue are extremely low for some observations. We will replace them with the median
imdb_df_new.loc[imdb_df_new['budget']<1000,'budget'] = imdb_df_new.loc[imdb_df_new['budget']>=1000, 'budget'].median()
imdb_df_new.loc[imdb_df_new['revenue']<1000,'revenue'] = imdb_df_new.loc[imdb_df_new['revenue']>=1000, 'revenue'].median()
```

```python
imdb_df_new.describe()
```

# Profit can be calculated as the difference between the revenue and budget

```python
# Create profit column
imdb_df_new['profit'] = imdb_df_new['revenue'] - imdb_df_new['budget']
```

```python
imdb_df_new.head()
```

```python
# Calculate maximum profit
max_profit = imdb_df_new['profit'].max()
```

```python
max_profit_movie_df = imdb_df_new[imdb_df_new['profit']==max_profit]
```

```python
max_profit_movie_df
```

```python
# Creating a list of crew members
crew_list = max_profit_movie_df.loc[1761, 'crew']
```

```python
# Get the names of directors and producers
for elem in crew_list:
  if elem['job'] == 'Producer':
    print(f"The Producer of the movie is {elem['name']}")
  if elem['job'] == 'Director':
    print(f"The Director of the movie is {elem['name']}")
```
